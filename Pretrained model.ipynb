###### Part I
##### importing the VGG16 and Create Model

```python
import os
import glob
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from keras.utils import load_img
import matplotlib.pyplot as plt
from keras.preprocessing.image import *
```

```python
model = VGG16(weights = 'imagenet')
```

```python
model.summary()
```

##### Loading the own image of ourselves

```python
#Load flower jpeg image from local and set target size to 224 x 224
img = tf.keras.utils.load_img('dataset/my/my1.jpg', target_size=(224,224))
```

```python
img =  tf.keras.utils.img_to_array(img)
img = np.expand_dims(img, axis = 0)
img
```

##### print prediction

```python
predicted = model.predict(img)
```

```python
predicted
```

```python

top_five_predict = tf.keras.applications.vgg16.decode_predictions(predicted,
                                                                  top=5)
top_five_predict
```
